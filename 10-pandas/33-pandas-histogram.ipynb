# 1) Pandas histogram

- histogram - graficke prevedenie distribucie numerickych udajov
- **hist()** - build-in f-cia, vkladame pole udajov ako parameter
- **bins** - optional parameter of hist() f-cie, rozsah hodnot, ktory reprezentuje skupinu udajov


```python
import pandas as pd
import matplotlib.pyplot as plt

# create a DataFrame
data = {
    "values": [
        18,
        21,
        22,
        25,
        28,
        30,
        32,
        34,
        35,
        36,
        38,
        40,
        42,
        45,
        50,
        55,
        60,
        65,
        70,
    ]
}
df = pd.DataFrame(data)

# plot a histogram
plt.hist(df, bins=10)
plt.show()
```

## 1.1) Customized histogram


```python
import pandas as pd
import matplotlib.pyplot as plt

# create a DataFrame with more data
data = {"values": [23, 45, 30, 50, 67, 35, 47, 62, 25, 58, 42, 36, 53, 68, 32]}
df = pd.DataFrame(data)

# plot a customized histogram
plt.hist(df, bins=7, edgecolor="black", color="green", alpha=0.7)
plt.title("Customized Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

## 1.2) Multiple histograms

- na porovnanie roznych datasetov


```python
import pandas as pd
import matplotlib.pyplot as plt

# Create two DataFrames with different datasets
data1 = {"values": [12, 15, 18, 22, 25, 27, 30, 33, 37, 40]}
data2 = {"values": [8, 10, 14, 20, 24, 28, 32, 36, 42, 45]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Plot two histograms side by side
plt.hist(df1, bins=6, edgecolor="black", alpha=0.7, label="Dataset 1")
plt.hist(
    df2,
    bins=6,
    edgecolor="black",
    alpha=0.7,
    label="Dataset 2",
    color="orange",
)
plt.title("Histogram Comparison")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
