## 1.1) Line plot

- **plot()**
- kind = "line"


```python
import pandas as pd
import matplotlib.pyplot as plt

car = ["Caterham", "Tesla", "Audi", "BMW", "Ford", "Jeep"]
weight = [0.48, 1.7, 2, 2, 2.3, 3]

# create a DataFrame
data = {"Car": car, "Weight": weight}
df = pd.DataFrame(data)

# plot using Pandas
df.plot(x="Car", y="Weight", kind="line", marker="o")
plt.xlabel("Car")
plt.ylabel("Weight")
plt.title("Car Weights")
plt.show()
```

## 1.2) Scatter plot

- **plot()**
- kind = "scatter"


```python
import pandas as pd
import matplotlib.pyplot as plt

car = ["Caterham", "Tesla", "Audi", "BMW", "Ford", "Jeep"]
weight = [0.48, 1.7, 2, 2, 2.3, 3]

# create a DataFrame
data = {"Car": car, "Weight": weight}
df = pd.DataFrame(data)

# scatter plot using Pandas
df.plot(x="Car", y="Weight", kind="scatter", marker="o", color="blue")
plt.xlabel("Car")
plt.ylabel("Weight")
plt.title("Car Weights (Scatter Plot)")
plt.grid(True)
plt.show()
```

## 1.3) Bar graph

- **plot()**
- kind = "bar"


```python
import pandas as pd
import matplotlib.pyplot as plt

car = ["Caterham", "Tesla", "Audi", "BMW", "Ford", "Jeep"]
weight = [0.48, 1.7, 2, 2, 2.3, 3]

# create a DataFrame
data = {"Car": car, "Weight": weight}
df = pd.DataFrame(data)

# bar graph using Pandas
df.plot(x="Car", y="Weight", kind="bar", color="blue")
plt.xlabel("Car")
plt.ylabel("Weight")
plt.title("Car Weights (Bar Graph)")
plt.tight_layout()  # function is used to ensure that the plot layout is adjusted properly
plt.show()
```

## 1.4) Histograms

- **plot()**
- kind = "hist"


```python
import pandas as pd
import matplotlib.pyplot as plt

weight = [0.48, 1.7, 2, 3]

# create a DataFrame
data = {"Weight": weight}
df = pd.DataFrame(data)

# histogram using Pandas
df.plot(kind="hist", bins=10, edgecolor="black", color="blue")
plt.show()
```
