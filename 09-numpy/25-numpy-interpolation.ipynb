# 1) Numpy interpolation

- interpolaciou urcujeme hodnotu funkcie v bodoch kde hodnoty nie su zname
- **interp()**

```python
import numpy as np

day = np.array([2, 4, 7])
gold_price = np.array([55, 58, 65])

# find the value of gold on day 3
day3_value = np.interp(3, day, gold_price)

print("The value of gold on day 3 is", day3_value)
```

## 1.2) Graph the interpolated values

```python
import numpy as np
import matplotlib.pyplot as plt

# arrays with data points
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 3, 4, 5, 7, 8])

# generate evenly (rovnomerne) spaced (rozmiestnene) values between
# the minimum and maximum x values
x_interp = np.linspace(x.min(), x.max(), 100)  # 100 je pocet bodov, je to optional
print(x_interp)

# interp() to interpolate y values
y_interp = np.interp(x_interp, x, y)

# plot the original data points and the interpolated values
plt.plot(x, y, "bo")
plt.plot(x_interp, y_interp, "r-")
```
