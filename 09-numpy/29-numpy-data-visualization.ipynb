# 1) Numpy data visualization

## 1.1) Line plot

**plot()**

```python
import numpy as np
import matplotlib.pyplot as plt

car = np.array(["Caterham", "Tesla", "Audi", "BMW", "Ford", "Jeep"])
weight = np.array([0.48, 1.7, 2, 2, 2.3, 3])

plt.plot(car, weight)
plt.show()
```

## 1.2) Scatter plot

- **scatter()**
- argumenty f-cie scatter():
    - **c** - farba bodu
    - **s**- velkost bodu

```python
import numpy as np
import matplotlib.pyplot as plt

car = np.array(["Caterham", "Tesla", "Audi", "BMW", "Ford", "Jeep"])
weight = np.array([0.48, 1.7, 2, 2, 2.3, 3])

# set colors and sizes
colors = np.array([0, 1, 2, 3, 4, 5])
sizes = np.array([20, 40, 60, 80, 100, 120])

plt.scatter(car, weight, c=colors, s=sizes)
plt.show()
```

## 1.3) Bar graphs

- **bar()**

```python
import matplotlib.pyplot as plt
import numpy as np

car = np.array(["Caterham", "Tesla", "Audi", "BMW", "Ford", "Jeep"])
weight = np.array([0.48, 1.7, 2, 2, 2.3, 3])

plt.bar(car, weight)
plt.title("Bar graph")
plt.show()
```

## 1.4) Histograms

- **hist()**

```python
import numpy as np
import matplotlib.pyplot as plt

weight = np.array([0.48, 1.7, 2, 3])

plt.hist(weight)
plt.show()
```
