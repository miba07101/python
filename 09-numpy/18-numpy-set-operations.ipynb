# 1) Numpy set operations

- set je subor unikatnych udajov
- set operacie:
    - union
    - intersection
    - difference
    - symetric difference
    - unique

## 1.1) Union

- syntax: `np.union1d()` jeho ekvivalent je `A ⋃ B` v set operaciach




```python
import numpy as np

A = np.array([1, 3, 5])
B = np.array([0, 2, 3])

print("A:\n", A)
print("B:\n", B)

# union of two arrays
print("Union:\n", np.union1d(A, B))
```

## 1.2) Intersection

- syntax: `np.intersect1d(A,B)` jeho ekvivalent je `A ⋂ B` v set operaciach


```python
import numpy as np

A = np.array([1, 3, 5])
B = np.array([0, 2, 3])

print("A:\n", A)
print("B:\n", B)

# intersection of two arrays
print("Intersection:\n", np.intersect1d(A, B))
```

## 1.3) Difference

- syntax: `np.setdiff1d(A,B)` jeho ekvivalent je `A - B` v set operaciach


```python
import numpy as np

A = np.array([1, 3, 5])
B = np.array([0, 2, 3])

print("A:\n", A)
print("B:\n", B)

# difference of two arrays
print("Difference:\n", np.setdiff1d(A, B))
```

## 1.4) Symmetric difference

- syntax: `np.setxor1d(A,B)`


```python
import numpy as np

A = np.array([1, 3, 5])
B = np.array([0, 2, 3])

print("A:\n", A)
print("B:\n", B)

# symmetric difference of two arrays
print("Symmetric diff:\n", np.setxor1d(A, B))
```

## 1.5) Unique values from array

- syntax: `np.unique()`
- vrati pole zatriedenych unikatnych hodnot

```python
import numpy as np

arr = np.array([1, 1, 2, 2, 4, 7, 7, 3, 5, 2, 5])
print("Array:", arr)

# unique values from array1
print("Unique values:", np.unique(arr))
```
