# 1) Perfect binary tree

- kazdy "internal" uzol ma presne 2 "child" uzly a posledne "leaf" uzly su na rovnakej urovni



- rekurzivne(spatne) mozme "perfect binary tree" dafinovat ako:
    1. ak samostany uzol nema ziadne "child" uzly, tak je to "perfect binary tree" o vyske (height) **h = 0**
    2. ak uzol ma **h > 0**, tak je to "perfect binary tree", ak oba jeho sub-stromy maju vysku **h - 1** a neprekryvaju sa

    

## 1.1) Predpoklady

1. perfect binary tree (PBT) o vyske **h** ma: 2**(h + 1) - 1 uzlov
2. PBT s **n** uzlami ma vysku: log(n + 1) - 1 = 0(ln(n))
3. PBT o vyske **h** ma: 2**h "leaf" uzlov
4. priemerna hlbka uzla je: 0(ln(n))

```python
# Checking if a binary tree is a perfect binary


# vytvorenie uzla
class Node:
    def __init__(self, data):
        self.data = data
        self.leftChild = None
        self.rightChild = None


# vypocet hlbky stromu
def calcDepth(node):
    if node is None:
        return 0
    left_depth = calcDepth(node.leftChild)
    right_depth = calcDepth(node.rightChild)

    return max(left_depth, right_depth) + 1


# kontrola ci je to PBT
def isPBT(root, d, level=0):
    # kontrola ci je prazdny
    if root is None:
        return True

    # kontrola ci su "leaf" uzly, resp. "child"
    if root.leftChild is None and root.rightChild is None:
        return d == level + 1  # ??? musim zistit co to je

    if root.leftChild is None and root.rightChild is None:
        return False

    return isPBT(root.leftChild, d, level + 1) and isPBT(root.rightChild, d, level + 1)


root = Node(1)
root.leftChild = Node(2)
root.rightChild = Node(3)
root.leftChild.leftChild = Node(4)
root.leftChild.rightChild = Node(5)
root.rightChild.leftChild = Node(6)
root.rightChild.rightChild = Node(7)

if isPBT(root, calcDepth(root)):
    print("The tree is a perfect binary tree")
else:
    print("The tree is not a perfect binary tree")
```
