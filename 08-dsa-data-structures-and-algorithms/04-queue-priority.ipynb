# 1) Priority queue

- specialny datovy typ fronty/radu, pri ktorom je kazdy prvok spojeny s prioritnou hodnotou. 
- s prvkami sa pracuje v poradi pridelenej priority.
- ak su prvky s rovnakou prioritou, prednost ma ten ktory bol pridany do radu skor.
- pr. nastavenia priority - cislo s vacsou hodnotou ma vyssiu prioritu, ale mozme nastavit prioritu lubovolne



## 1.1) Difference between priority and normal (simple) queue

- v normalnom rade/fronte uplatnujeme pravidlo FIFO. V "priority" rade su prvky odstarnovane na zaklade priority. Prvky s vyssou prioritou su odstranovane prv.

## 1.2) Implementation

- pomocou:
    - array (pola)
    - linked list (prepojene listy)
    - heap data (hromada dat)
    - binary search tree
