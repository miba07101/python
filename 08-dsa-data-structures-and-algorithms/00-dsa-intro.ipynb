# 1) Data structures and types

- data structure is a storage to store and organize data
- data structure is a collection of **data types** arranged in a specific order
- ** data structure types:**
    - linear
    - non-linear
- **! Programming is all about data structures and algorithms. Data structures are used to hold data while algorithms are used to solve the problem using that data!**

## 1.1) Linear data structures

- elements are arranged in sequence one after the other
- **types:**
    - array (pole) - napr. list
    - stack (zasobnik) - LIFO principle of stored elements (LIFO - last in, first out)
    - queue (fronta) - FIFO - first in, first out
    - linked list (prepojeny zoznam) - ellemnts are connected through a series of nodes

## 1.2) Non-linear data structures

- elements are not arranged in any sequence
- elements are arranged in **hierarchical** maner where one element is connected to one or more elements
- **types:**
    - graph - each element/node is called **vertex (vrchol)**. Vertex is connected to other vertices through **edges**
    - trees - difference is that there can only be one edge between 2 vertices

# 2) Asymptotic analysis

- asymptotic notations - su matematicke zapisy pouzivane na popis casu chodu algoritmu, ak vstupna hodnota sa blizi k urcitej hodnote alebo limitnej hodnote
- 3 hlavne asymptoticke zapisy:
    - Big-O
    - Omega
    - Theta

## 2.1) Big-O notation

- predstavuje hornu hranicu casu chodu algoritmu
- dava najhorsi pripad zlozitosti algoritmu



## 2.2) Omega notation

- predstavuje dolnu hranicu casu chodu algoritmu
- predstavuje najlepsi pripad zlozitosti algoritmu



## 2.3) Theta notation

- predstavuje hornu a dolnu hranicu casu chodu algoritmu
- pouziva sa na analyzu priemernej zlozitosti algoritmu



# 3) Divide and conquer algorithm

- na riesenie zlozitych uloh postupom:
    - divide - rozbitie problemu na mensie sub-problemy
    - conquer - riesenie sub-problemov
    - combine - skombinovanie sub-rieseni na dosiahnutie pozadovaneho vysledku
- pouziva sa **rekurzia**
- najcastejsia aplikacia:
    - binary search
    - merge sort
    - quick sort
    - atd..
- vypocet casovej narocnosti:
    - pomocou **master theorem** T(n) - aT(n/b) + f(n)
        - n - size of input
        - a - number of sub-problems in the recursion
        - n/b - size of each sub-problem. all sub-problems are assumed to have the same size
        - f(n) - cost of the work done outside the recursive call
    - riesenie pre **merge sort**:
        - a = 2 (each time a problem is divided into 2 sub-problems)
        - n/b = n/2 (size of each sub-problem is half of the input)
        - f(n)- (time taken to divide the problem and merging the sub-problems)
        - T(n/2) = 0(n log n) - (to vyjde z master theorem rovnice)
        - T(n/2) - 2T(n/2) + f(n) --> 0(n log n)
