# 1) Stack data structures

- **LIFO** - last in, first out - putt element in is called **push** and removing element is called **pop**



- basic operations of stack:
    - **push** - add element to the top of stack
    - **pop** - remove an element from the top of a stack
    - **IsEmpty** - check if the stack is empty
    - **IsFull** - check if the stack is full
    - **Peek** - get the value of the top element without removing it

## 1.2) Working of stack data structure

1) **TOP** - pointer - pouziva sa na sledovanie horneho prvku v zasobniku/stacku
2) pri inicializacii/vytvarani zasobnika, nastavime pointer TOP na hodnotu -1, aby sme porovnanim mohli zistit ci je zasobnik prazdny
3) pridanim prvku zmenime hodnotu pointera TOP a umiestnime novy prvok na miesto kde ukazuje pointer
4) pri odobrati prvku znizime hodnotu pointeru TOP
5) pred pridanim prvku, skontrolujeme, ci zasobnik uz nie je plny 
6) pred odoberanim prvku, skontrolujeme, ci zasobnik nie je prazdny

```python
# Stack implementation



# creating a Stack

def create_stack():

    stack = []  # create empty list
    return stack



# check if stack is empty

def check_empty(stack):

    return len(stack) == 0



# add item to stack

def push(stack, item):

    stack.append(item)

    print("pushed item", item)



# remove item from stack

def pop(stack):

    if check_empty(stack):

        return "stack is empty"

    return stack.pop()



# stack initialization

stack = create_stack()

push(stack, 1)

push(stack, 2)

push(stack, 3)

push(stack, 4)
print(stack)
pop(stack)

print("stack after poping element", stack)
```

## 1.3) Application of stack data structure

- reverse a word - put all letters in a stack and pop them out. Because of LIFO order of stack, you will get letters in reverse order

```python
# reverse word using stack
word = "zelenina"

# najednoduchsie riesenie pomocou "slicing"
reverse_word = word[::-1]
print(reverse_word)


# riesenie pomocou stack
def reverse(word):
    stack = []
    reverse_stack = []
    for letter in word:
        stack.append(letter)

    for _ in range(0, len(stack)):
        if len(stack) == 0:
            print("stack is empty")
        else:
            reverse_stack.append(stack.pop())
    return reverse_stack


reverse(word)
```
