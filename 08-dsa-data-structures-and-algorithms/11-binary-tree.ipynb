# 1) Binary tree

- typ datovej struktury, kde kazdy "parent" uzol ma najviac 2 "child" uzly
- kazdy uzol obsahuje:
    - data
    - adresu na lavy "child" uzol
    - adresu na pravy "child" uzol



## 1.1) Types

### 1. Full binary tree

- kazdy "parent" uzol ma 2 alebo ziaden "child" uzol


### 2. Perfect binary tree

- kazdy "internal" uzol ma presne 2 "child" uzly a posledne "leaf" uzly su na rovnakej urovni



### 3. Complete binary tree

- je ako **full binary tree** ale:
    - kazda uroven musi byt uplne zaplnena
    - vsetky posledne "leaf" uzly sa musia ??naklonit (must lean towards)?? dolava
    - posledny "leaf" uzol nemusi mat praveho surodenca (right sibling)
    
    

### 4. Degenerate or Pathological tree

- datova struktura, kde je vzdy len 1 "child" uzol ci uz nalavo alebo napravo



### 5. Skewed binary tree

- je to v podstate degenerate/pathological tree, ale "child" uzly su bud nalavo alebo napravo
- pozname teda 2 typy:
    - left-skewed binary tree
    - right-skewed binary tree
    
    

### 6. Balanced binary tree

- rozdiel medzi "height" (vyskou) laveho sub-stromu a praveho sub-stromu pre kazdy uzol je 0 alebo 1.



## 1.2) Binary tree representation

- uzol binary tree pozostava z:
    - data
    - 2 "pointers/address" na dalsie uzly, resp. strukturu rovnakeho typu



```python
# binary tree
# v podstate implementacia z 10-tree-traversal-algorithm.ipynb


# vytvorenie uzla
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None  # pointer na lavy uzol
        self.right = None  # pointer na pravy uzol

    # traversal f-cie
    def inorder(self):
        # left traversal
        if self.left:
            self.left.inorder()
        # root
        print(self.data, end=" ")
        # right traversal
        if self.right:
            self.right.inorder()

    def preorder(self):
        # root
        print(self.data, end=" ")
        # left traversal
        if self.left:
            self.left.preorder()
        # right traversal
        if self.right:
            self.right.preorder()

    def postorder(self):
        # left traversal
        if self.left:
            self.left.postorder()
        # right traversal
        if self.right:
            self.right.postorder()
        # root
        print(self.data, end=" ")


# vytvorenie instancie
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)

print("In order Traversal: ", end="")
root.inorder()
print("\nPre order Traversal: ", end="")
root.preorder()
print("\nPost order Traversal: ", end="")
root.postorder()
```
