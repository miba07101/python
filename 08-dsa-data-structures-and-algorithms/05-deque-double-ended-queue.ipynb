# 1) Deque data structures

- pridavanie alebo odoberanie prvkov moze byt vykonane na oboch koncoch/stranach FRONT aj REAR.



- typy:
    - **input restricted deque** - pridavanie je obmedzene na jeden koniec ale odoberanie moze byt z oboch koncov
    - **output restricted deque** - odoberanie je obmedzene na jeden koniec ale pridavanie moze byt na oboch koncoch

## 1.1) Operations on a deque

1) vytvorenie prazdneho pola
    - vytvor pole o velkosti -n (-k) prvkov
    - nastav FRONT = -1 a REAR = 0
2) Insert at FRONT
    - overim ci pole nie je plne
    - ak pole je plne `(front == 0 && rear == n - 1) || (front == rear + 1)`, nemozem pridat prvok do pola/radu/fronty
    - ak je pole prazdne, tak posuniem poziciu FRONT = 0 a pridam prvok na tuto poziciu `array[front]`
    - ak FRONT = 0, tak zmen FRONT = n - 1 (posledny index v poli)
    - pridaj prvok na poziciu FRONT
    - zniz FRONT o -1
3) Insert at REAR
    - overim ci pole nie je plne
    - ak je plne, tak nemozem pridat prvok
    - ak je pole prazdne, tak REAR = 0, pridam prvok `array[rear]`
    - ak REAR = n - 1 (posledny prvok), tak zmen REAR = 0
    - zvysim REAR o +1
    - pridam prvok na poziciu REAR
4) Delete from the FRONT
    - skontrolujem ci je pole prazdne
    - ak je prazdne FRONT = -1, tak nemozem odobrat prvok
    - ak ma pole iba jeden prvok `FRONT=REAR`, tak nastavim FRONT a REAR = -1
    - ak je FRONT = n - 1 (posledny prvok), tak FRONT = 0, inac zvysim hodnotu FRONT o +1
5) Delete from the REAR
    - skontrolujem ci je pole prazdne
    - ak je prazdne FRONT = -1, tak nemozem odobrat prvok
    - ak ma pole iba jeden prvok `FRONT=REAR`, tak nastavim FRONT a REAR = -1
    - ak je REAR = 0 (prvy prvok), tak REAR = n - 1 (posledny prvok), inac znizim hodnotu REAR o -1
6) Check empty
    - je prazdne pole, ak FRONT = -1
7) Check full
    - pole je plne ak `(FRONT == 0 and REAR == n - 1) or FRONT = REAR + 1`

```python
# deque implementation
# trochu zvlastne oproti tomu co je v popise


class Deque:

    # konstruktor
    def __init__(self):
        self.array = []

    # je prazdne
    def isEmpty(self):
        # return aby mi vypisalo True alebo False
        return self.array == []

    # pridat z FRONT
    def addFront(self, item):
        self.array.insert(0, item)

    # pridat z REAR
    def addRear(self, item):
        self.array.append(item)

    # odobrat z FRONT
    def removeFront(self):
        self.array.pop(0)

    # odobrat z REAR
    def removeRear(self):
        # ak je return mozem vypisat prvok, ktory som odstranil
        return self.array.pop()

    # velkost pola
    def sizeArray(self):
        # ak je reutrn, mozem vypisat hodnoty velkosti pola
        return len(self.array)

    # print pola
    def printDeque(self):
        print(self.array)


# inicializacia instancie triedy
deque = Deque()

deque.isEmpty()
deque.addFront(5)
deque.printDeque()
deque.addFront(1)
deque.addFront(2)
deque.printDeque()
deque.addRear(3)
deque.printDeque()
deque.removeFront()
deque.printDeque()
print(deque.removeRear())
deque.printDeque()
print(deque.sizeArray())

# vypis pola aj bez funkcie printArray()
print(deque.array)
```
