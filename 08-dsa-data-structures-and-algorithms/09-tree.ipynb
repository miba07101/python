# 1) Tree data structure

- nelinearna hierarchicka datova struktura
- je zlozena z "uzlov (nodes)" spojenymi "hranami (edges)"
- pouziva sa pretoze pri "linearnych" datovych strukturach (polia, linked list, stack, queue) casova narocnost rastie umerne s velkostou udajov.
- terminologia:
    - **node (uzol)**
        - obsahuje "key alebo value" a "pointer" na jeho "child" uzol
        - posledny uzol kazdej cesty sa nazyva **leaf (list)** alebo **external** node. Neobsahuje "pointer"
        - uzol, ktory ma "child" uzol sa vola **internal** uzol
    - **edge (hrana)**
        - spojenie medzi dvoma uzlami
        
    
    
    - **root**
        - prvy (hlavny) uzol
    - **height of a node**
        - vyska uzla je pocet hran k poslednemu "leaf" uzlu (je to najdlhsia cesta od uzla k leaf uzlu)
    - **depth of a node**
        - je pocet hran od "root" uzla k danemu uzlu
    - **height of a tree**
        - je vyska "root" uzla alebo hlbka (depth) najhlbsieho uzla

    
    
    - **degree of a node**
        - celkovy pocet vetiev uzla
    - **forest**
        - kolekcia nespojenych (disjoint) stromov (trees)
        - mozem vytvorit aj "odrezanim" root uzla

    
    

## 1.1) Types

- binary tree
- binary search tree (BST)
- AVL tree
- B-tree

## 1.2) Aplikacia

- binary search tree sa pouziva na rychle overenie ci prvok je pritomny v "set" alebo nie
- heap je druh stromu pouzivany na "heap sort"
- databazy pouzivaju B-tree a T-tree modely
- compiler pouziva strom na overenie syntaxu kazdeho programu, ktory napiseme
