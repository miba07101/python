# 1) Full binary tree

- kazdy "parent" uzol ma 2 alebo 0 "child" uzlov
- niekedy sa vola aj **proper binary tree**



## 1.1) Predpoklady

- i - pocet vnutornych "internal" uzlov
- n - celkovy pocet uzlov
- l - pocet poslednych uzlov "leaves"
-  - pocet urovni "levels"
- teda plati:
    1. pocet leaves = i + 1
    2. celkovy pocet uzlov = 2i + 1
    3. pocet internal uzlov = (n - 1) / 2
    4. pocet leaves = (n + 1) / 2
    5. celkovy pocet uzlov = 2l - 1
    6. pocet internal uzlov = l - 1
    7. pocet leaves najviac = 2**( - 1)

```python
# Checking if a binary tree is a full binary tree in Python


# vytvorenie uzla
class Node:
    def __init__(self, data):
        self.data = data
        self.leftChild = None
        self.rightChild = None


# checking full binary tree
def isFullTree(root):
    # prazdny strom
    if root is None:
        return True

    # kontrola ci "child" uzly su pritomne
    if root.leftChild is None and root.rightChild is None:
        return True

    if root.leftChild is not None and root.rightChild is not None:
        # rekurzia f-cie
        return isFullTree(root.leftChild) and isFullTree(root.rightChild)

    return False


# inicializacia instancie triedy Node
root = Node(1)
root.rightChild = Node(3)
root.leftChild = Node(2)

root.leftChild.leftChild = Node(4)
root.leftChild.rightChild = Node(5)
root.leftChild.rightChild.leftChild = Node(6)
root.leftChild.rightChild.rightChild = Node(7)

if isFullTree(root):
    print("The tree is a full binary tree")
else:
    print("The tree is not a full binary tree")
```
