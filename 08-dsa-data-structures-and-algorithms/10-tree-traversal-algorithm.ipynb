# 1) Tree traversal - inorder, preorder and postorder

- traversing (prechadzat) stromom znamena navstivit kazdy uzol v strome
- kazdy strom je kombinaciou:
    - uzlu v ktorom su ulozene data
    - dvoch sub-stromov



- v zavislosti od sposobu ako chceme navstivit kazdy uzol v strome pozname tieto typy "traversal":
    - inorder
    - preorder
    - postorder

## 1.1) Inorder traversal

1) najprv navstivim kazdy uzol v "lavom" sub-strome
2) potom "root" uzol
3) nakonie vsetky uzly v "pravom" sub-strome

- postup:


    - prejdeme v poradi: lavy sub-strom, root, pravy sub-strom
    - ulozime do **stack (nadrz)**


    - prejdeme sub-strom ulozeny na vrchu nadrze (opat v poradi: lavy sub-strom, root, pravy sub-strom)


    - kedze uzol "5" uz nema ziadne sub-stromy traversal je ukonceny
    - kedze stack je LIFO (last in, first out) dostaneme **inorder traversal** vo forme: `5 -> 12 -> 6 -> 1 -> 9`

- !Important: nemusime vytvarat "stack", kedze rekurziou zabezpecime spravne poradie


## 1.2) Preorder traversal

1) najprv "root" uzol
2) potom vsetky uzly v "lavom" sub-strome
3) nakoniec vsetky uzly v "pravom" sub-strome

## 1.3) Postorder traversal

1) vsetky uzly v "lavom" sub-strome
2) potom vsetky uzly v "pravom" sub-strome
3) nakonie "root" uzol

```python
# Tree traversal


# vytvorenie uzla
class Node:
    # konstruktor
    def __init__(self, item):
        self.left = None  # child uzol na lavo
        self.right = None  # child uzol na pravo
        self.value = item


# traversal f-cie
def inorder(root):
    if root:
        # traverse left
        inorder(root.left)
        # travers root
        print(root.value, "->", end=" ")
        # travers right
        inorder(root.right)


def preorder(root):
    if root:
        # travers root
        print(root.value, "->", end=" ")
        # traverse left
        preorder(root.left)
        # traverse right
        preorder(root.right)


def postorder(root):
    if root:
        # traverse left
        postorder(root.left)
        # traverse right
        postorder(root.right)
        # travers root
        print(root.value, "->", end=" ")


# inicializacia instancie triedy Node
root = Node(1)
root.left = Node(12)
root.right = Node(9)
root.left.left = Node(5)
root.left.right = Node(6)

print("\nInorder traversal ")
inorder(root)
print("\nPreorder traversal ")
preorder(root)
print("\nPostorder traversal ")
postorder(root)
```
