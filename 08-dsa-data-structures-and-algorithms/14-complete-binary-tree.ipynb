## 1.2) How to create Complete binary tree

1. vyber prvy prvok z pola ako root uzol (pocet prvkov na level-I:1 ??)


2. vloz druhy prvok ako lavy "child" uzol a treti prvok ako pravy "child" uzol (pocet prvkov na level-II:2)


3. vloz dalsie prvky (stvrty a piaty) ako "child" uzly laveho uzla z level-II. dalsie dva uzly prirad pravemu uzlu z level-II. (pocet prvkov na level-III:4)

4. opakuj po posledny prvok v poli


## 1.3) Relationship between array indexes and tree element (Vztah medzi indexom pola a uzlom stromu)

- ak index prvku z pola je **i**, tak prvok s indexom **2i + 1** bude lavy "child" uzol a prvok s indexom **2i + 2** bude pravy "child" uzol
- "parent" uzol prvku s indexom **i** urcime ako: **(i - 1) / 2**

```python
# Checking if a binary tree is a complete binary tree


class Node:
    def __init__(self, data):
        self.data = data
        self.leftChild = None
        self.rightChild = None


# vypocet poctu uzlov
def count_nodes(root):
    if root is None:
        return 0
    # pomocou rekurzie scitam pocet uzlov
    # 1 je pretoze prvy je root uzol
    return 1 + count_nodes(root.leftChild) + count_nodes(root.rightChild)


# kontrola ci je complete binary tree
def isCBT(root, index, numNodes):
    if root is None:
        return True

    if index >= numNodes:
        return False

    return (isCBT(root.leftChild, 2 * index + 1, numNodes)) and (
        isCBT(root.rightChild, 2 * index + 2, numNodes)
    )


# inicializacia
root = Node(1)
root.leftChild = Node(2)
root.rightChild = Node(3)
root.leftChild.leftChild = Node(4)
root.leftChild.rightChild = Node(5)
root.rightChild.leftChild = Node(6)

node_count = count_nodes(root)
print(node_count)

index = 0

if isCBT(root, index, node_count):
    print("The tree is a complete binary tree")
else:
    print("The tree is not a complete binary tree")
```
